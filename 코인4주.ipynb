# Columns and rows together

- 행으로 subset 만들기
- 그 후 열로 subset 만든다.

위 단계는 문제를 해결하는데 있어 좋은방법입니다. 데이터 프레임을 검사하여 각 단계에서 찾고 있는 결과를 얻을 수 있습니다.



````python
# 30살 이하의 고객들의 location은 어디인가요?
# 행으로 subset을 먼저 추출해냅니다.
df = cars[cars['Age'] < 30]

# 그 후에 열로 subset을 만들어줍니다.
df2 = df['Location']
df2
````

## Excel results


````python
# profit이 최소 3000이상인 age 컬럼과 vehicle type컬럼의 subset을 구하시오
df = cars[cars['Profit']>= 3000]
df2 = df[['Age','Vehicle-Type']]
df2
````

## Excel results


````python
# 이전 구매기록이 1개보다 많은 고객들의 평균 profit은? 


# 우선 Previous > 1인 subset을 만듭니다 
# 그 후에 profit컬럼의 평균을 구합니다.

df = cars[cars['Previous']>1]
df['Profit'].mean()      
# 위 메소드는 오직 결과만 표시합니다.

# 만약 위 평균값을 저장하고 싶다면 변수를 만들어 아래와 같이 저장해주면 됩니다.
#frequent_cust = df['Profit'].mean()
````
